# Fitting(Training) Model (50pts)

```python
MIN_USER_RATINGS = 35
DELETE_RATING_COUNT = 15

###############################################################################
# TODO : Complete the below function in way that, getting ratings array, creates
# two same size arrays as train and validation, and for the users in ratings array
# that have at least "MIN_USER_RATINGS" observed ratings, choose "DELETE_RATING_COUNT"
# of these ratings, make them zero in train table and put them in validation table
# return train and validation tables
###############################################################################


def train_test_split(ratings):

    train = np.copy(ratings)
    validation = np.zeros(ratings.shape)
    
    for i in range(ratings.shape[0]):
        if np.count_nonzero(ratings[i,:]) >= MIN_USER_RATINGS:
            arg_rating = np.random.choice(np.nonzero(ratings[i,:])[0],size=DELETE_RATING_COUNT,replace=False)
            validation[i,arg_rating] = train[i,arg_rating]
            train[i,arg_rating] = 0

    
    return train, validation
```

```python
train, val = train_test_split(ratings)
```

```python
###############################################################################
# TODO : Complete below function in a way to return sqrt value of "mean_squared_error"
# between non-zero(observed) values of ground_truth and prediction. Note that you sould
# care about observed values of ground_truth table not prediction!
###############################################################################


def rmse(prediction, ground_truth):
    
    prediction = prediction[np.nonzero(ground_truth)].flatten()
    ground_truth = ground_truth[np.nonzero(ground_truth)].flatten()

    return sqrt(mean_squared_error(prediction, ground_truth))
```

```python
class Recommender:
    
    def __init__(self, n_epochs=200, n_latent_features=3, lmbda=0.1, learning_rate=0.001, C=1):
        self.n_epochs = n_epochs
        self.n_latent_features = n_latent_features
        self.lmbda = lmbda
        self.learning_rate = learning_rate
        self.C = C
        self.train_error = []
        self.valid_error = []

    def predictions(self, P, Q):
        return np.dot(P.T,Q)
  
    def fit(self, X_train, X_val):

        Q =  self.C*np.random.randn(self.n_latent_features, X_train.shape[1])
        P =  self.C*np.random.randn(self.n_latent_features, X_train.shape[0])

        user, item = np.nonzero(X_train)
        
        for n in range(self.n_epochs):

            self.train_error.append(rmse(self.predictions(P, Q), X_train))
            self.valid_error.append(rmse(self.predictions(P, Q), X_val))
            
            for i, j in zip(user, item):

                error = X_train[i, j] - self.predictions(P[:,i], Q[:,j])
                Q[:, j] = Q[:, j] + self.learning_rate * (error * P[:, i] - self.lmbda * Q[:, j])                
                P[:, i] = P[:, i] + self.learning_rate * (error * Q[:, j] - self.lmbda * P[:, i])

        self.Q = Q
        self.P = P
            
        return self

    def predict(self, X_train, user_index):
        
        pre_index = np.where(X_train[user_index, :] == 0)[0]
        y_pre = self.predictions(self.P, self.Q)
        pre = y_pre[user_index, pre_index].flatten()
        
        return pre
```

```python
#creating recommender object
recommender = Recommender().fit(train, val)
```

```python
###############################################################################
# TODO : Plot the loss of train and loss
###############################################################################

print('RMSE of Training Data = ', recommender.train_error[-1])
print('RMSE of Validation Data = ', recommender.valid_error[-1])

plt.plot(np.arange(recommender.n_epochs), recommender.train_error, color='r', label='Training Loss')
plt.plot(np.arange(recommender.n_epochs), recommender.valid_error, color='b', label='Validation Loss')
plt.xlabel('number of epochs')
plt.ylabel('RMSE')
plt.legend()
plt.show()
```

```python
def create_artist_ratings(artists_df, artists_index, ratings, n=10):
  artist_ids = ratings_df.columns[artists_index]
  artist_ratings = pd.DataFrame(data=dict(artistId=artist_ids, rating=ratings))
  top_n_artists = artist_ratings.sort_values("rating", ascending=False).head(n)
  
  artist_recommendations = artists_df[artists_df.id.isin(top_n_artists.artistId)].reset_index(drop=True)
  artist_recommendations['rating'] = pd.Series(top_n_artists.rating.values)
  return artist_recommendations.sort_values("rating", ascending=False)
```
